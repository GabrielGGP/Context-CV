```python
import cv2
import numpy as np
from pycocotools.coco import COCO
```

```python
def maskForPicture(picture, segmentation):
    img = cv2.imread(picture)
    # create a mask 0 will be black and 255 will be true.
    mask = np.zeros(img.shape[:2], np.uint8)
    
    #Translate the position dict to a list of 2D points
    pairs = []
    for i in range(0,len(segmentation),2):
        #x,y = str(pos).split('.')
        #pairs.append([int(x), int(y)])
        print(i)
        pairs.append([round(segmentation[i]), round(segmentation[i+1])])
        
    # Define the polygon points
    pts = np.array(pairs)
    #pts = np.array([[100, 100], [100, 50], [50, 50]])

    # Draw the polygon on the mask
    cv2.fillPoly(mask, [pts], (255, 255, 255))

    #mask[100:250, 150:450] = 255

    # compute the bitwise AND using the mask
    masked_img = cv2.bitwise_and(img,img,mask = mask)

    # display the mask, and the output image
    cv2.imshow('Mask',mask)
    cv2.waitKey(0)
    cv2.imshow('Masked Image',masked_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
object_annotation_path = "Dataset/annotations/object_annotations/instances_train2017.json"
object_annotations = COCO(annotation_file=object_annotation_path)
object_ids = object_annotations.getCatIds()
object_ids
```

```python
zebraId = 24
object_annotations.getImgIds(catIds=[zebraId])
```

```python
#34 is picture of zebra. Picture of zebra.
pictureToLoad = 106935
pictureAnnotations = object_annotations.loadAnns(object_annotations.getAnnIds(pictureToLoad, iscrowd=None))
```

```python
#Itterating through all the array like annotations.
for j in pictureAnnotations:
    if j['category_id'] == zebraId:
        print("Found the zebra")
        #We wound the zebra, now we call the maskForPicture to get the mask.
        #First we find where the picture is
        img_info = object_annotations.loadImgs([pictureToLoad])[0]
        img_file_name = img_info["file_name"]
        img_full_path = f"./Dataset/train2017/{img_file_name}"
        print(f"{j['segmentation']}")
        print(f"/Dataset/train2017/{img_file_name}")
        #j['segmentation'] is an array of x,y positions separated by a (.) used to mask an object.
        #TODO: Add support for multiple of the same object.
        maskForPicture(img_full_path, j['segmentation'][0])
```
